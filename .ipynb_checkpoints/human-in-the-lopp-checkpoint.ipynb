### Human in the loop

In previous chapters we saw **autonomous** agents by turning human_input_mode="NEVER".



Actually, this human input mode sits between the auto reply functionality (see above pic)

We will learn the usage of it now.

### Human input modes

1. NEVER - disabled
2. TERMINATE (default) - HI requested when 'terminate condition' is met
3. ALWAYS - HI is always requested. HI can either
    - terminate chat
    - provide feedback and continue
    - skip (to trigger auto reply)

    In this mode, any other termination conditions are ignored.

#### HI='NEVER'

Use when you want the agents to act fully autonomously.

<u>Below is an example given in the tutorial picked as is to illustrate 'NEVER' i.e. fully autonomous</u>

*Note how the system message in this case of 2 bots is composed from a first person perspective of the other agent.*

```python
import os

from autogen import ConversableAgent

agent_with_number = ConversableAgent(
    "agent_with_number",
    system_message="You are playing a game of guess-my-number. You have the "
    "number 53 in your mind, and I will try to guess it. "
    "If I guess too high, say 'too high', if I guess too low, say 'too low'. ",
    llm_config={"config_list": [{"model": "gpt-4o-mini", "api_key": os.environ["OPENAI_API_KEY"]}]},
    is_termination_msg=lambda msg: "53" in msg["content"],  # terminate if the number is guessed by the other agent
    human_input_mode="NEVER",  # never ask for human input
)

agent_guess_number = ConversableAgent(
    "agent_guess_number",
    system_message="I have a number in my mind, and you will try to guess it. "
    "If I say 'too high', you should guess a lower number. If I say 'too low', "
    "you should guess a higher number. ",
    llm_config={"config_list": [{"model": "gpt-4o-mini", "api_key": os.environ["OPENAI_API_KEY"]}]},
    human_input_mode="NEVER",
)

result = agent_with_number.initiate_chat(
    agent_guess_number,
    message="I have a number between 1 and 100. Guess it!",
)
```

#### HI='ALWAYS'

As defined above.

Example below.

```python
human_proxy = ConversableAgent(
    "human_proxy",
    llm_config=False,  # no LLM used for human proxy
    human_input_mode="ALWAYS",  # always ask for human input
)

# Start a chat with the agent with number with an initial guess.
result = human_proxy.initiate_chat(
    agent_with_number,  # this is the same agent with the number as before
    message="10",
)
```

#### HI='TERMINATE' (default)

When this is the HI mode, on attainment of termination condition, HI is requested.

Example below where guessing bot terminates after 2 replies and then HI kicks in.

As we already know, human can either
1. skip (trigger auto-reply)
2. terminate
3. intercept and human brain reply

```python
from autogen import ConversableAgent
import os

agent_with_number = ConversableAgent(
    "agent_with_number",
    system_message="You are playing a game of guess-my-number. You have the "
    "number 53 in your mind, and I will try to guess it. "
    "If I guess too high, say 'too high', if I guess too low, say 'too low'. ",
    llm_config={"config_list": [{"model": "gpt-4o-mini", "api_key": os.environ["OPENAI_API_KEY"]}]},
    is_termination_msg=lambda msg: "53" in msg["content"],  # terminate if the number is guessed by the other agent
)

agent_guess_number = ConversableAgent(
    "agent_guess_number",
    system_message="I have a number in my mind, and you will try to guess it. "
    "If I say 'too high', you should guess a lower number. If I say 'too low', "
    "you should guess a higher number. ",
    llm_config={"config_list": [{"model": "gpt-4o-mini", "api_key": os.environ["OPENAI_API_KEY"]}]},
    human_input_mode="TERMINATE",
    max_consecutive_auto_reply=2
)

result = agent_with_number.initiate_chat(
    agent_guess_number,
    message="I have a number between 1 and 100. Guess it!",
)
```
