## <u>Miscellaneous functions</u>

Learning and documenting other miscellaneous functions for completeness

### <u>autogen_agentchat.ui.Console</u>

Asynchronous function that renders live AgentChat interactions in our terminal.

#### <u>Breakdown</u>

Msg signature



1. Consumes the message stream (async generators/iterators) generated by **run_stream** or **on_messages_stream** from AgentChat agents.
2. Renders messages real-time to your console - in a chat-style.
3. Returns a final object (everything together) at the end
   - TaskResult for run_stream
   - Response for on_messages_stream
